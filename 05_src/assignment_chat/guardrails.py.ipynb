```python
BLOCKED_TOPICS = [
    "cat",
    "cats",
    "dog",
    "dogs",
    "horoscope",
    "zodiac",
    "taylor swift"
]


def check_guardrails(message):

    text = message.lower()

    for topic in BLOCKED_TOPICS:
        if topic in text:
            return False, "Sorry, I cannot answer questions about that topic."

    forbidden = [
        "system prompt",
        "show me your instructions",
        "ignore previous instructions",
        "modify your instructions",
        "reveal your prompt"
    ]

    for item in forbidden:
        if item in text:
            return False, "Sorry, I cannot reveal or modify my system instructions."

    return True, ""
```
